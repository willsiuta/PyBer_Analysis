### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to check these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

city_data_df.isnull().sum()
ride_data_df.isnull().sum()
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride = pyber_data_df.groupby(by='type')['fare'].sum() / pyber_data_df.groupby(by='type')['ride_id'].count()
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = pyber_data_df.groupby(by='type')['fare'].sum() / pyber_data_df.groupby(by='type')['driver_count'].sum()
average_fare_per_driver
```

```python
# Step 1 data
total_rides = pyber_data_df.groupby(by='type')['ride_id'].count()

# Step 2 data
total_driver_count = city_data_df.groupby(by='type')['driver_count'].sum()


# Step 3 data
total_fares = pyber_data_df.groupby(by='type')['fare'].sum()

# Step 4 data
average_fare_per_ride = total_fares / total_rides

# Step 5 data
average_fare_per_driver = total_fares / total_driver_count

# Create the summary DataFrame
summary_df = pd.DataFrame({'Total Rides': total_rides,
                         'Total Drivers': total_driver_count,
                         'Total Fares': total_fares,
                         'Average Fare per Ride': average_fare_per_ride,
                         'Average Fare per Driver': average_fare_per_driver})

# format the last 3 columns
summary_df["Total Fares"] = summary_df["Total Fares"].map('${:,.2f}'.format)
summary_df["Average Fare per Ride"] = summary_df["Average Fare per Ride"].map('${:,.2f}'.format)
summary_df["Average Fare per Driver"] = summary_df["Average Fare per Driver"].map('${:,.2f}'.format)
summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
summary_df.index.name = None
summary_df
```

```python
pyber_data_df_datetype_group_pivot = pyber_data_df_datetype_group_no_index.pivot(index='date', columns='type', values='fare')
pyber_data_df_datetype_group_pivot
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-28'.
pyber_2019_Jan_Apr = pyber_data_df_datetype_group_pivot.loc['2019-01-01':'2019-04-28']
pyber_2019_Jan_Apr
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
pyber_2019_Jan_Apr.index = pd.to_datetime(pyber_2019_Jan_Apr.index)
pyber_2019_Jan_Apr
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
pyber_2019_Jan_Apr.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fare = pyber_2019_Jan_Apr.resample('W').sum()
weekly_fare
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

ax = weekly_fare.plot(figsize = (20,6))
# Add a title 
ax.set_title("Total Fare by City Type")
#  Add a x- and y-axis label.
ax.set_xlabel("Month")
ax.set_ylabel("Fare ($USD)")
plt.savefig("Analysis/Challenge_fare_summary.png")
plt.show()
```
